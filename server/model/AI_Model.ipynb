```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data_df = pd.read_csv("/content/full_2024_stats.csv")
```

```python
data_df
data_opps = data_df.copy()
data_df.drop(columns=['opponent'], inplace=True)
```

```python
data_opps
```

```python
#encoding text columns
from sklearn.preprocessing import LabelEncoder
df_encoded = pd.DataFrame.copy(data_df)
df_encoded.drop(columns=['player_id', 'name', 'games'], inplace=True)
le = LabelEncoder()
for i in df_encoded.columns:
  if df_encoded[i].dtype==object:
    le.fit(df_encoded[i])
    np.save(f'{df_encoded[i]._name} classes.npy', le.classes_)
    df_encoded[i]=le.transform(df_encoded[i])


for column in df_encoded.columns:
  df_encoded[column] = df_encoded[column].astype(float)

df_encoded
```

```python
df_slim = df_encoded.loc[~((df_encoded['xp'] == 0) & (df_encoded['pass_TD'] == 0) & (df_encoded['xp_miss'] == 0)
                            & (df_encoded['fgs'] == 0) & (df_encoded['fg_miss'] == 0) & (df_encoded['fumbles'] == 0)
                            & (df_encoded['ints'] == 0) & (df_encoded['pass_yds'] == 0) & (df_encoded['rush_yds'] == 0)
                            & (df_encoded['rush_TD'] == 0) & (df_encoded['rec_yds'] == 0) & (df_encoded['receptions'] == 0)
                            & (df_encoded['rec_TD'] == 0))]

df_slim.reset_index(drop=True, inplace=True)

df_slim
```

```python
df_slim.columns
```

```python
data = df_slim[['team', 'weight', 'height', 'year', 'position']].values.tolist()
labels = df_slim[['pass_TD', 'xp', 'xp_miss', 'fgs', 'fg_miss', 'fumbles',
                  'ints', 'pass_yds', 'rush_yds',
                  'rush_TD', 'rec_yds', 'rec_TD', 'receptions']].values.tolist()

```

```python
#Splitting data into test-train
from sklearn.model_selection import train_test_split

training_data, test_data, training_labels, test_labels = train_test_split(data, labels, test_size=0.2)
```

```python
# Create the MLPRegressor model
model = MLPRegressor(
    hidden_layer_sizes=(100, 50, 25),  # Two hidden layers with 64 and 32 neurons
    activation='relu',           # ReLU activation function
    solver='adam',               # Adam optimizer
    #max_iter=50000000,                # Maximum number of iterations
    #random_state=42
)

# Fit the model to the training data
model.fit(training_data, training_labels)

# Predict on the test set
y_pred = model.predict(test_data)

# Evaluate the model
mse = mean_squared_error(test_labels, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
test_QB = df_test.values.reshape(1, -1)
test_results = model.predict(test_QB)

test_results.tolist()
#df_slim[['xp', 'xp_miss', 'fgs', 'fg_miss', 'fumbles',
#                  'ints', 'pass_yds', 'rush_yds',
#                  'rush_TD', 'rec_yds', 'rec_TD', 'receptions']].values.tolist()

test_dict = {}



test_dict['pass_TD'] = round(test_results[0][0], 1)
test_dict['xp'] = round(test_results[0][1], 1)
test_dict['xp_miss'] = round(test_results[0][2], 1)
test_dict['fgs'] = round(test_results[0][3], 1)
test_dict['fg_miss'] = round(test_results[0][4], 1)
test_dict['fumbles'] = round(test_results[0][5], 1)
test_dict['ints'] = round(test_results[0][6], 1)
test_dict['pass_yds'] = round(test_results[0][7], 1)
test_dict['rush_yds'] = round(test_results[0][8], 1)
test_dict['rush_TD'] = round(test_results[0][9], 1)
test_dict['rec_yds'] = round(test_results[0][10], 1)
test_dict['rec_TD'] = round(test_results[0][11], 1)
test_dict['receptions'] = round(test_results[0][12], 1)

for i in test_dict:
  test_dict[i] = abs(test_dict[i])
  if test_dict[i] <= 0.1:
    test_dict[i] = 0
  if i == 'receptions':
    test_dict[i] = round(test_dict[i], 0)
    if test_dict[i] == 0:
      test_dict["rec_yds"] = 0
      test_dict["rec_TD"] = 0



test_dict
```
