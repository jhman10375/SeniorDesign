```python
class fbGame():
    def __init__(self, game_id: int, home_id: int, home_team: str, away_team: str, start_date: str):
        self.game_id = game_id
        self.home_id = home_id
        self.home_team = home_team
        self.away_team = away_team
        self.start_date = start_date

    def __str__(self):
        return f"game_id={self.game_id} home_id={self.home_id} home_team={self.home_team} away_team={self.away_team} start_date={self.start_date}"
```
